# Importando Bibliotecas

```python
from bs4 import BeautifulSoup as bs
import requests as rq
import re
```

### Coleta de dados da Wikipedia

```python
scraping_filolau = rq.get("https://pt.wikipedia.org/wiki/Filolau_de_Crotona")
filolau = bs(scraping_filolau.content)
print(filolau.prettify())
```

### Coletando todos os 'h2'

```python
h2 = filolau.find_all("h2")
print(h2)
```

### Em forma de lista (For)

```python
for h2_list in h2:
  print(h2_list)
```

```python
for h2_list in h2:
  print(h2_list.text)
```

### Coletando 'H1'

```python
h1 = filolau.find_all("h1")
print(h1)
```

```python
for h1_list in h1:
  print(h1_list.text)
```

### Coletando todos os 'h1 e h2'

```python
h2_h1 = filolau.find_all(["h2","h1"])
for h2_h1_list in h2_h1:
  print(h2_h1_list.text)
```

### Coletando os 'p'

```python
p = filolau.find('p')
for p_list in p :
  print(p_list.text)
```

### Coletando 'a'

```python
a = filolau.find('a')
for a_list in a :
  print(a_list.text)
```
